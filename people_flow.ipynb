```python
import madina as md
import madina.una.tools as una

import multiprocessing
print(multiprocessing.cpu_count())
```

```python
Bancho = md.Zonal()
```

```python
Bancho.load_layer("Buildings",r"C:\Users\86135\OneDrive - Georgia Institute of Technology\桌面\上完的课程\Tokyo Studio\Bancho\Madina_Bancho\Bancho_origin.geojson")
Bancho.load_layer("Sidewalk",r"C:\Users\86135\OneDrive - Georgia Institute of Technology\桌面\Madina_Bancho\Pedestrian_Rds_new.geojson")
Bancho.load_layer("Clinic",r"C:\Users\86135\OneDrive - Georgia Institute of Technology\桌面\Madina_Bancho\Healthcare_POIs.geojson")
```

```python
Bancho.describe()
```

```python
Bancho["Sidewalk"].gdf["weighted"] = Bancho["Sidewalk"].gdf["Shape_Length"] * (1+1*Bancho["Sidewalk"].gdf["tanθ"])
```

```python
df= Bancho["Sidewalk"].gdf
df
```

```python
Bancho.create_street_network( source_layer= "Sidewalk", node_snapping_tolerance= 0.1,weight_attribute="weighted")
```

```python
Bancho.insert_node(layer_name='Buildings',label="origin",weight_attribute='pop_adj')
Bancho.insert_node(layer_name='Clinic',label="destination")
```

```python
Bancho.create_graph()
```

```python
una.betweenness(
    Bancho,
    search_radius = 500,
    detour_ratio=1.05,
    decay=True,
    decay_method="exponent",
    beta=0.001,
    turn_penalty=False,
    closest_destination=False,
    num_cores=32,
    save_betweenness_as="Home_to_Clinic")
```
