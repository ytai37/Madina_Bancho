```python
Bancho = md.Zonal()
```

```python
Bancho.load_layer("Buildings",r"C:\Users\86135\OneDrive - Georgia Institute of Technology\桌面\上完的课程\Tokyo Studio\Bancho\Madina_Bancho\Bancho_origin.geojson")
Bancho.load_layer("Sidewalk",r"C:\Users\86135\OneDrive - Georgia Institute of Technology\桌面\上完的课程\Tokyo Studio\Bancho\Pedestrian_Rds_new.geojson")
Bancho.load_layer("Park",r"C:\Users\86135\OneDrive - Georgia Institute of Technology\桌面\上完的课程\Tokyo Studio\Bancho\Madina_Bancho\park.geojson")
```

```python
Bancho.describe()
```

```python
df= Bancho["Sidewalk"].gdf
df
```

```python
Bancho.create_street_network( source_layer= "Sidewalk", node_snapping_tolerance= 0.1,weight_attribute="Weighted")
```

```python
Bancho.insert_node(layer_name='Buildings',label="origin",weight_attribute='pop_adj')
Bancho.insert_node(layer_name='Park',label="destination")
```

```python
Bancho.create_graph()
```

```python
una.betweenness(
    Bancho,
    search_radius = 500,
    detour_ratio=1.05,
    decay=True,
    decay_method="exponent",
    beta=0.001,
    turn_penalty=False,
    closest_destination=False,
    num_cores=32,
    save_betweenness_as="Home_to_Park")
```
